#### _Importing neccessary libraries and creating DataFrames by reading the 2 datasets_

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

life_expectancy_df = pd.read_csv('life_expectancy.csv')
countries_of_the_world_df = pd.read_csv('countries of the world.csv')

life_expectancy_df
```

### _Creating a function which aggregates the data from the life_expectancy.csv dataset by calculating the average values for the collumns: Hepatitis B, Measles, Polio, Diphtheria, HIV/AIDS, BMI, Alcohol, Life expectancy, and GDP for each unique country in the dataset and creates a new DataFrame for this aggregated data._

The function processes the life_expectancy_df dataset, which contains yearly data for various health and economic indicators of countries over a 16-year period, calculates and returns the 16-year averages for these indicators for each country.

```python
def aggregate_life_expectancy_dataset():
    countries_list, hepatitis_b_list, measles_list, polio_list, diphteria_list, hiv_aids_list, bmi_list, alchohol_list, life_expectancy_list, gdp_list = [], [], [], [], [], [], [], [], [], []

    current_country = ""
    index = 1
    hepatitis_b_container, measles_container, polio_container, diphteria_container, hiv_aids_container, bmi_container, alchohol_container, life_expectancy_container, gdp_container = 0, 0, 0, 0, 0, 0, 0, 0, 0

    life_expectancy_df["Hepatitis B"].replace(np.nan, life_expectancy_df["Hepatitis B"].mean(), inplace  = True)
    life_expectancy_df["Measles"].replace(np.nan, life_expectancy_df["Measles"].mean(), inplace  = True)
    life_expectancy_df["Polio"].replace(np.nan, life_expectancy_df["Polio"].mean(), inplace  = True)
    life_expectancy_df["Diphtheria"].replace(np.nan, life_expectancy_df["Diphtheria"].mean(), inplace  = True)
    life_expectancy_df["HIV/AIDS"].replace(np.nan, life_expectancy_df["HIV/AIDS"].mean(), inplace  = True)
    life_expectancy_df["BMI"].replace(np.nan, life_expectancy_df["BMI"].mean(), inplace  = True)
    life_expectancy_df["Alcohol"].replace(np.nan, life_expectancy_df["Alcohol"].mean(), inplace  = True)
    life_expectancy_df["Life expectancy"].replace(np.nan, life_expectancy_df["Life expectancy"].mean(), inplace  = True)
    life_expectancy_df["GDP"].replace(np.nan, life_expectancy_df["GDP"].mean(), inplace  = True)

    for row_name, row_data in life_expectancy_df.iterrows():
        if(index % 16 == 0):
            hepatitis_b_list.append(round((hepatitis_b_container/16),2))
            measles_list.append(round((measles_container/16),2))
            polio_list.append(round((polio_container/16),2))
            diphteria_list.append(round((diphteria_container/16),2))
            hiv_aids_list.append(round((hiv_aids_container/16),2))
            bmi_list.append(round((bmi_container/16),2))
            alchohol_list.append(round((alchohol_container/16),2))
            life_expectancy_list.append(round((life_expectancy_container/16),2))
            gdp_list.append(round((gdp_container/16),2))
            hepatitis_b_container, measles_container, polio_container, diphteria_container, hiv_aids_container, bmi_container, alchohol_container, life_expectancy_container, gdp_container = 0, 0, 0, 0, 0, 0, 0, 0, 0
        if(row_data['Country'] not in countries_list):
            current_country = row_data['Country']
            countries_list.append(row_data['Country'])
        if(index % 16 != 0):
            hepatitis_b_container += float(row_data['Hepatitis B'])
            measles_container += float(row_data['Measles'])
            polio_container += float(row_data['Polio'])
            diphteria_container += float(row_data['Diphtheria'])
            hiv_aids_container += float(row_data['HIV/AIDS'])
            bmi_container += float(row_data['BMI'])
            alchohol_container += float(row_data['Alcohol'])
            life_expectancy_container += float(row_data['Life expectancy'])
            gdp_container += float(row_data['GDP'])
        index += 1
        
    filtered_average_life_expectancy_df = pd.DataFrame(
        {
            'Country': countries_list,
            '(AVG) Hepatitis B (2000-2015)': hepatitis_b_list,
            '(AVG) Measles (2000-2015)': measles_list,
            '(AVG) Polio (2000-2015)': polio_list,
            '(AVG) Diphtheria (2000-2015)': diphteria_list,
            '(AVG) HIV/AIDS (2000-2015)': hiv_aids_list,
            '(AVG) BMI (2000-2015)': bmi_list,
            '(AVG) Alcohol (2000-2015)': alchohol_list,
            '(AVG) Life expectancy (2000-2015)': life_expectancy_list,
            '(AVG) GDP (2000-2015)': gdp_list
        }
    )
    
    return filtered_average_life_expectancy_df   
```

### _Merge the countries of the world dataset with the aggregated dataset_

The function aggregates the life expectancy dataset, cleans country names in both datasets, merges the two datasets based on country names, removes entries with any missing values and returs a consolidated dataset with complete information for each country.

```python
def merge_and_remove_inconsitent_data():
    aggregated_life_expectancy_df = aggregate_life_expectancy_dataset()

    countries_of_the_world_df['Country'] = countries_of_the_world_df['Country'].str.strip()
    aggregated_life_expectancy_df['Country'] = aggregated_life_expectancy_df['Country'].str.strip()

    merged_df = countries_of_the_world_df.merge(aggregated_life_expectancy_df, on='Country', how = 'outer')
    merged_df = merged_df.dropna()
    
    return merged_df
```

```python
final_df = merge_and_remove_inconsitent_data()
final_df
```

#### _Visualization of the average GDP per capita and the average death rate for each region._

This function generates a bar chart that shows the average GDP per capita and the average death rate for each region of the final merged Dataframe.

```python
def gdp_vs_deathrate():
    final_df["GDP ($ per capita)"] = final_df["GDP ($ per capita)"].astype(str).str.replace(',', '').astype(float)
    final_df["Deathrate"] = final_df["Deathrate"].astype(str).str.replace(',', '').astype(float)
    
    grouped = final_df.groupby("Region").agg({
        "GDP ($ per capita)": "mean",
        "Deathrate": "mean"
    })
    
    gdp_per_capita = grouped["GDP ($ per capita)"]
    deathrate = grouped["Deathrate"]

    plt.figure(figsize=(18, 10))
    plt.bar(grouped.index, gdp_per_capita, label="GDP ($ per capita)", color="blue")
    plt.bar(grouped.index, deathrate, label="Deathrate", color="red", alpha=0.7)
    plt.xlabel("Region")
    plt.ylabel("GDP")
    plt.title("GDP per Capita vs. Deathrate by Region")
    plt.xticks(rotation=90)
    plt.legend()
    plt.tight_layout()
    
    plt.show()
```

#### _Visualization of the relationship between the average GDP per capita and the average birthrate for all different regions_

This function produces a bar chart that creates a visual comparison between the average GDP per capita and the average birthrate across all different regions in the final merged DataFrame.

```python
def gdp_vs_birthrate():
    final_df["GDP ($ per capita)"] = final_df["GDP ($ per capita)"].astype(str).str.replace(',', '').astype(float)
    final_df["Birthrate"] = final_df["Birthrate"].astype(str).str.replace(',', '').astype(float)
    
    grouped = final_df.groupby("Region").agg({
        "GDP ($ per capita)": "mean",
        "Birthrate": "mean"
    })
    
    gdp_per_capita = grouped["GDP ($ per capita)"]
    birthrate = grouped["Birthrate"]

    plt.figure(figsize=(18, 10))
    plt.bar(grouped.index, gdp_per_capita, label="GDP ($ per capita)", color="blue")
    plt.bar(grouped.index, birthrate, label="Birthrate", color="red", alpha=0.7)
    plt.xlabel("Region")
    plt.ylabel("Birthrate")
    plt.title("GDP per Capita vs. Birthrate by Region")
    plt.xticks(rotation=90)
    plt.legend()
    plt.tight_layout()
    
    plt.show()
```

#### _Visualization of the average prevalence of the diseases: Polio, Diphteria, HIV/AIDS and Hepatitis B across all  diferent regions for the years 2000 - 2015_

This function creates a visual representation of how the average cases of four diseases vary across different regions over a fifteen-year span.

```python
def correlation_between_diseases():
    grouped = final_df.groupby("Region").agg({
        "(AVG) Polio (2000-2015)": "mean",
        "(AVG) Diphtheria (2000-2015)": "mean",
        "(AVG) HIV/AIDS (2000-2015)": "mean",
        "(AVG) Hepatitis B (2000-2015)": "mean"
    })

    fig, ax = plt.subplots(figsize=(20,8))
    plt.xticks(rotation=90)

    ax.plot(grouped.index, grouped["(AVG) Polio (2000-2015)"], label="Polio")
    ax.plot(grouped.index, grouped["(AVG) Diphtheria (2000-2015)"], label="Diphtheria")
    ax.plot(grouped.index, grouped["(AVG) HIV/AIDS (2000-2015)"], label="HIV/AIDS")
    ax.plot(grouped.index, grouped["(AVG) Hepatitis B (2000-2015)"], label="Hepatitis B")

    ax.set_title("Correlation between different diseases among different regions (2000-2015)", fontstyle='italic', fontsize=16, loc='left')
    ax.set_xlabel('Region', fontweight='bold')
    ax.set_ylabel('Cases', fontweight='bold')
    ax.legend(fontsize=10, title_fontsize=30, loc='lower right')
    plt.tight_layout()
    plt.show()
```

#### _Visualization of the average GDP and life expectancy of different regions for the years 2000 - 2015_

This function provides a side-by-side comparison, through a bar chart, of the average GDP (scaled down by a factor of 10 for better visualization) and the average life expectancy across all different regions over the specified time period.

```python
def correlation_gdp_life_expectancy():
    grouped = final_df.groupby("Region").agg({
        "(AVG) GDP (2000-2015)": "mean",
        "(AVG) Life expectancy (2000-2015)": "mean",
    })

    avg_gdp = grouped["(AVG) GDP (2000-2015)"].astype(str).str.replace(',', '').astype(float) / 10
    life_expectancy = grouped["(AVG) Life expectancy (2000-2015)"].astype(str).str.replace(',', '').astype(float)

    plt.figure(figsize=(18, 10))
    plt.bar(grouped.index, avg_gdp, label="GDP", color="blue")
    plt.bar(grouped.index, life_expectancy, label="Life expectancy", color="red", alpha=0.7)
    plt.xlabel("Region")
    plt.ylabel("GDP (Divided by 10)")
    plt.title("(AVG) GDP (2000-2015) vs. (AVG) Life expectancy (2000-2015)")
    plt.xticks(rotation=90)
    plt.legend()
    plt.tight_layout()
    
    plt.show()
```

#### _Visualization of the average GDP ($ per capita) and the average coastline-to-area ration of all different regions_

This function produces a line plot visualization of the correlation between the average DGP per capita and the coastline-to-area ratio accross different all different regions.

```python
def correlation_gdp_coastline():
    final_df["GDP ($ per capita)"] = final_df["GDP ($ per capita)"].astype(str).str.replace(',', '').astype(float)
    final_df["Coastline (coast/area ratio)"] = final_df["Coastline (coast/area ratio)"].astype(str).str.replace(',', '').astype(float)
    
    grouped = final_df.groupby("Region").agg({
        "Coastline (coast/area ratio)": "mean",
        "GDP ($ per capita)": "mean"
    })

    fig, ax = plt.subplots(figsize=(20,8))
    plt.xticks(rotation=90)

    ax.plot(grouped.index, grouped["GDP ($ per capita)"] / 10, label="GDP ($ per capita)")
    ax.plot(grouped.index, grouped["Coastline (coast/area ratio)"], label="Coastline (coast/area ratio)")

    ax.set_title("Correlation between GDP and Coastline", fontstyle='italic', fontsize=16, loc='left')
    ax.legend(fontsize=10, title_fontsize=30, loc='lower right')
    plt.tight_layout()
    plt.show()
```

#### _Visualization of the correlation between GDP per capita and literacy rates for all different regions_

The function creates a line plot in order to visualize the relationship between the average GDP per capita and literacy rates across all different regions.

```python
def correlation_gdp_literacy():
    final_df["GDP ($ per capita)"] = final_df["GDP ($ per capita)"].astype(str).str.replace(',', '').astype(float)
    final_df["Literacy (%)"] = final_df["Literacy (%)"].astype(str).str.replace(',', '').astype(float)
    
    grouped = final_df.groupby("Region").agg({
        "Literacy (%)": "mean",
        "GDP ($ per capita)": "mean"
    })

    fig, ax = plt.subplots(figsize=(20,8))
    plt.xticks(rotation=90)

    ax.plot(grouped.index, grouped["GDP ($ per capita)"] / 10, label="GDP ($ per capita)")
    ax.plot(grouped.index, grouped["Literacy (%)"], label="Literacy (%)")

    ax.set_title("Correlation between GDP and Literacy (%)", fontstyle='italic', fontsize=16, loc='left')
    ax.legend(fontsize=10, title_fontsize=30, loc='lower right')
    plt.tight_layout()
    plt.show()
```
